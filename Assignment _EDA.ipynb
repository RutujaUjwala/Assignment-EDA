```python
#11) Identify and remove outliers in the km_driven column using the IQR method
```

```python
# 11 ans)

#To check outliers>> distplot, boxplot
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.distplot(df['km_driven'],kde=True)
plt.title('Km_driven by Bike_distplot')

plt.subplot(1,2,2)
sns.boxplot(data=df['km_driven'])
plt.title('Km_driven by Bike_boxplot')
plt.show()
```

```python
'''here the  distribution is right skwewd  and the outlier are present in right 
side.
As we able to see boxplot also showing the outliers.
To overcome outliers we are using following steps.'''
```

```python
# Step 1: Calculate Q1, Q3, and IQR for km_driven

Q1=df['km_driven'].quantile(0.25) #the 25th percentile (Q1) 
Q3=df['km_driven'].quantile(0.75)#75th percentile (Q3)
#identify outliers and gives insights
IQR=Q3-Q1
print(f'iqr_km_driven :{IQR}')

# Step 2: Define the outlier thresholds
lower_bound_km_driven=Q1-1.5*IQR
upper_bound_km_driven=Q1+1.5*IQR

print(f"Lower Bound: {lower_bound_km_driven}")
print(f"Upper Bound: {upper_bound_km_driven}")

# Step 3: Identify outliers
outliers = df[(df['km_driven'] < lower_bound_km_driven) | (df['km_driven'] > upper_bound_km_driven)]
print(f"Number of outliers in km_driven: {len(outliers)}")

# Step 4: Remove outliers

df_cleaned = df[(df['km_driven'] >= lower_bound_km_driven) & (df['km_driven'] <= upper_bound_km_driven)]

print(f"Number of rows after removing outliers: {len(df_cleaned)}")
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.distplot(df_cleaned['km_driven'],kde=True)
plt.title('Km_driven by Bike_distplot')

plt.subplot(1,2,2)
sns.boxplot(data=df_cleaned['km_driven'])
plt.title('Km_driven by Bike_boxplot')
plt.show()
```
